**Heart Disease Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('heart_data/heart.csv')
```

 Dataset Loading and Initial Exploration

```python
print(df.head(5))
print(df.tail(5))
print(df.shape)
```

Dataset Structure and Missing Value Analysis

```python
print(df.columns)
print(df.info())
print(df.isnull().sum())
```

Categorical Data Analysis

```python
ptnt_heart_ds = df[df.target==1]
ptnt_no_heart_ds = df[df.target==0]
print(ptnt_heart_ds.shape)
print(ptnt_no_heart_ds.shape)

male_ptnt = df[df.sex == 1]
female_ptnt = df[df.sex == 0]
print(male_ptnt.shape)
print(female_ptnt.shape)
```

 Data Selection Using iloc

```python
first_tens = df.head(10)
print(first_tens[['age','sex','chol']])
print(df.iloc[20:31, 0:5])
```

 Data Filtering Using Conditions

```python
ptnt_older_than_50 = df[df.age>50]
ptnt_hd_chol_gt_240 = df[(df.target==1) & (df.chol>240)]
ptnt_both_count = (ptnt_hd_chol_gt_240['age']>50).sum()
print(ptnt_both_count)
```
